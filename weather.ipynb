```python
import requests
import datetime as dt 

BASE_URL = 'http://api.openweathermap.org/data/2.5/forecast?'
API_KEY = str(open('api_key.txt', 'r').read())
CITY = 'Stuttgart'
url = BASE_URL + "appid=" + API_KEY + "&q=" + CITY

response = requests.get(url).json()
print(len(response))
print(response)
```

```python
def get_weather_data(response):
    """ Die Funktion wird dazu verwendet , die verschiedene Wetter-Datenpunkte 
    einer gegebenen Stadt auszulesen 
        
    Args:
        data (dict): Dictionnary of weather params for the last 5 days 
    """
    # in list is weather , wind , dt_txt ,main
    #in main(dict) is temp_max , temp_min , feels_like , temp , pressure and humidity
    #in Wind(dict) is speed
    # in weather (list(dict)) is main , description and Icon 
    print(f'weather Prevision for {CITY} :')
    data_dict = response['list']
    if type(data_dict) is list :
        for i in data_dict :
            dt_txt = i['dt_txt']
            temp_min = round(float(i['main']['temp_min']) - 272.15 , 3)
            temp_max = round(float (i['main']['temp_max']) - 272.15 , 3)
            feels_like = round(float(i['main']['feels_like']) - 272.15 ,3 )
            pressure = float(i['main']['pressure']) #--> bar
            humidity = i['main']['humidity'] #--> prozentangabe
            wind = i['wind']['speed'] #--> m/s
            if type(i['weather']) is list and len(i['weather']) >=1 :
                description_set = dict()
                j = 0
                for k in i['weather']:
                    j = j + 1 
                    main_description = k['main']
                    description = k['description']
                    tuple_description = (main_description , description)
                    description_set[j] = tuple_description
        #print Ausgabe 
            print('--------------------------------------------------')
            print(dt_txt)
            print('###################################################')
            keys = list(description_set.keys())
            for key in keys :
                print(f'main description: {description_set[key][0]}')
                print(f'main description: {description_set[key][1]}')
            print(f'minimale Temperatur: {temp_min} °C')
            print(f'maximale Temperatur: {temp_max} °C')
            print(f'feels_like/gefühlt: {feels_like} °C')
            print(f'Pressure / Druck: {pressure} bar')
            print(f'humidity / Luftfeuchtigkeit: {humidity} %')
            print(f'wind speed / Luftgeschwindigkeit: {wind} m/s')
get_weather_data(dict(response))           
```
